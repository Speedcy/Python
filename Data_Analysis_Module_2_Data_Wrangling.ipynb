## Gérer les valeurs manquantes

```python
import pandas as pd
url = "automobile.csv"
df=pd.read_csv(url)
```

### Supprimer les lignes comportant des Nan

```python
df.dropna(subset=["price"],axis=0,inplace=True)
```

### Remplacer les valeur manquantes (ici '?') par la valeur moyenne de la série

```python
df.head(5)
```

Calcul de la valeur moyenne 

```python
mean=df["normalized-losses"][df["normalized-losses"]!='?'].astype(int).mean()
```

Remplacement de la valeur inconnue par la valeur moyenne

```python
df["normalized-losses"].replace('?',mean,inplace=True)
df.head(5)
```

## Data Binning - Séparer les données en plusieurs classes

Création de 3 classes de prix

```python
import numpy as np
df_price=df["price"][df["price"]!='?'].astype(int)
bins=np.linspace(min(df_price),max(df_price),4)
print(bins)
```

```python
group_names=["Low","Medium","High"]
df["price-binned"]=pd.cut(df_price,bins,labels=group_names,include_lowest=True)
```
